```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy
import os
#build figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
#rectangle builder
data = []
l,w,h = 16,16,16
shape = [l,w,h]
diml = range(l)
dimw = range(w)
dimh = range(h)
for l in diml:
    for w in dimw:
        for h in dimh:
            data.append([l,w,h])
```

```python
#visual
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
#minecraft: [x,y,z]
xs,ys,zs = [],[],[]
for i in data:
    xs = i[0]
    ys = i[1]
    zs = i[2]
    ax.scatter3D(xs,ys,zs,c='black')   
```

```python
#block states code, maped to 0-air 1-dirt (or other material)
hexdump = [0]*4096
for i in data:
    offset = 0
    x,y,z = i[0],i[2],i[1]
    if x % 2 == 0:
        offset += x+1
    else:
        offset += x-1
    offset += z*16
    offset += y*256
    hexdump[offset]=1
```

```python
#creates usable data for import
lints = []
for i in range(256):
    hexList = []
    hexStr = []
    for l in range(16):
        hexList.append(hexdump[i*16+l])
    if hexList[0]!=0:
        hexList = numpy.trim_zeros(hexList, trim='b')
    hexStr = ''.join(str(e) for e in hexList)
    lint = int(hexStr, 16)
    lints.append(lint)
```
